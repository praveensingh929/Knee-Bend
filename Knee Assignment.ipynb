```python
import cv2
import mediapipe as mp
import time
import math

 
mpDraw = mp.solutions.drawing_utils
mpPose = mp.solutions.pose
pose = mpPose.Pose(static_image_mode=False,
    model_complexity=2,
    min_tracking_confidence=0.7,
    min_detection_confidence=0.7)

count = 0
dir = 0
done=0
cap = cv2.VideoCapture('KneeBendVideo.mp4')
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

writer = cv2.VideoWriter('final kneesas3.mp4', cv2.VideoWriter_fourcc(*'mp4v'),23, (width, height))
 

pTime = 0
while True:

    success, img = cap.read()
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    results = pose.process(imgRGB)

    if results.pose_landmarks:
        
        for id, lm in enumerate(results.pose_landmarks.landmark):
            h, w, c = img.shape
            if(id==25 or id==23 or id==27):
                print(id, lm)
                lmList = []
                cx, cy = int(lm.x * w), int(lm.y * h)
                lmList.append([id, cx, cy])
                if(id==23):
                    x1=cx
                    y1=cy
                if(id==25):
                    x2=cx
                    y2=cy
                if(id==27):
                    x3=cx
                    y3=cy                                    

                print(lmList)
                
                

 
    cTime = time.time()
    fps = 1 / (cTime - pTime)
    pTime = cTime
    #print(x1,x2,x3)
    angle = math.degrees(math.atan2(y3 - y2, x3 - x2) -math.atan2(y1 - y2, x1 - x2))
    if angle < 0:
        angle += 360
    print(angle)
    
    cv2.line(img, (x1, y1), (x2, y2), (255, 255, 255), 3)
    cv2.line(img, (x3, y3), (x2, y2), (255, 255, 255), 3)
    cv2.circle(img, (x1, y1), 10, (0, 0, 255), cv2.FILLED)
    cv2.circle(img, (x1, y1), 15, (0, 0, 255), 2)
    cv2.circle(img, (x2, y2), 10, (0, 0, 255), cv2.FILLED)
    cv2.circle(img, (x2, y2), 15, (0, 0, 255), 2)
    cv2.circle(img, (x3, y3), 10, (0, 0, 255), cv2.FILLED)
    cv2.circle(img, (x3, y3), 15, (0, 0, 255), 2)
    cv2.putText(img, str(int(angle)), (x2 - 50, y2 + 50),
    cv2.FONT_HERSHEY_PLAIN, 2, (0, 0, 255), 2)
    
 

    if (angle<140):
        dir=1

    if(angle>140 and dir==1 and done==1):
        count+=1
        dir=0
        done=0
    
  
    
    if(dir==1 and angle<139 and angle>120):
        
        curr=time.time()

        
    if(dir==1 and angle<120):
        prev=time.time()
        now=int(prev-curr)
        now=now/2
        now=now/2
        

        cv2.putText(img, str(int(now)), (70, 120), cv2.FONT_HERSHEY_PLAIN, 3,
            (255, 0, 0), 3)
        if(now>0 and now<8): 
            cv2.putText(img,'KEEP HOLDING', (70, 150), cv2.FONT_HERSHEY_PLAIN, 3,
            (255, 0, 0), 3)
            
        if(now>8):
            done=1
            cv2.putText(img,'WELL DONE', (70, 150), cv2.FONT_HERSHEY_PLAIN, 3,
            (255, 0, 0), 3)
            
    

    cv2.putText(img, str(int(count)), (70, 50), cv2.FONT_HERSHEY_PLAIN, 3,
    (255, 0, 0), 3)
    
    writer.write(img)


    cv2.imshow("Image", img)
    if cv2.waitKey(1) == ord('q'):
        cap.release()
        cv2.destroyAllWindows()
        break
        
        
```
